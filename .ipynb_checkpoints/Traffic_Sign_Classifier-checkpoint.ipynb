## Step 2: Design and Test a Model Architecture

### Pre-process the Data Set
Although **grayscale** was so simple, it actually made the result better and training faster. But since I noticed some images were too dark and some were bright, I thought I should balance it and believed it would help my model work better.

I chose `CLAHE` for pre-processing because it could normalize the brightness according to the areas' brightness, which meant it could probably make dark areas brighter or bright areas darker.

```python
def CLAHE(img, tileGridSize=(3,3)):
    img = img.reshape((32,32,3))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=tileGridSize)
    return clahe.apply(img).reshape(-1, 32*32)
```

```python
X_train_normed = X_train.reshape(-1, 32*32*3)
X_train_normed = np.apply_along_axis(CLAHE, 1, X_train_normed, (3,3)).astype(np.float32).reshape(-1, 32, 32, 1)

X_valid_normed = X_valid.reshape(-1, 32*32*3)
X_valid_normed = np.apply_along_axis(CLAHE, 1, X_valid_normed, (3,3)).astype(np.float32).reshape(-1, 32, 32, 1)

X_test_normed = X_test.reshape(-1, 32*32*3)
X_test_normed = np.apply_along_axis(CLAHE, 1, X_test_normed, (3,3)).astype(np.float32).reshape(-1, 32, 32, 1)
```

### Model Architecture
I built a deeper neural network than LeNet with:
* **3 conv2d** layer (16, 32, and 64 filters) with **3 max pooling** layers (2 x 2)
* **3 mlp** hidden layers (312, 256, and 128 neurons), with `ReLu` activation function
* **Dropout** layers (0.8, 0.6, and 0.4 keep_prob) for each mlp layers, just after the activation functions

```python
def LeNet(x):
    mu = 0
    sigma = 0.1
    
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5,5,1,16), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(16))
    conv1 = tf.nn.conv2d(x, conv1_W, strides=[1,1,1,1], padding = 'SAME') + conv1_b
    conv1 = tf.nn.relu(conv1)
    conv1 = tf.nn.max_pool(conv1, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')
    
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5,5,16,32), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(32))
    conv2 = tf.nn.conv2d(conv1, conv2_W, strides=[1,1,1,1], padding = 'SAME') + conv2_b
    conv2 = tf.nn.relu(conv2)
    conv2 = tf.nn.max_pool(conv2, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')
    
    conv3_W = tf.Variable(tf.truncated_normal(shape=(5,5,32,64), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(64))
    conv3 = tf.nn.conv2d(conv2, conv3_W, strides=[1,1,1,1], padding = 'SAME') + conv3_b
    conv3 = tf.nn.relu(conv3)
    conv3 = tf.nn.max_pool(conv3, ksize=[1,2,2,1], strides=[1,2,2,1], padding='VALID')
    
    fc0 = flatten(conv3)

    fc1_W = tf.Variable(tf.truncated_normal(shape=(np.int(fc0.shape[1]),312), mean=mu, stddev=sigma))
    fc1_b = tf.Variable(tf.zeros(312))
    fc1 = tf.matmul(fc0, fc1_W) + fc1_b
    fc1 = tf.nn.relu(fc1)
    do1 = tf.nn.dropout(fc1, 0.8)
    
    fc2_W = tf.Variable(tf.truncated_normal(shape=(312,256), mean=mu, stddev=sigma))
    fc2_b = tf.Variable(tf.zeros(256))
    fc2 = tf.matmul(do1, fc2_W) + fc2_b
    fc2 = tf.nn.relu(fc2) 
    do2 = tf.nn.dropout(fc2, 0.6)
    
    fc3_W = tf.Variable(tf.truncated_normal(shape=(256,128), mean=mu, stddev=sigma))
    fc3_b = tf.Variable(tf.zeros(128))
    fc3 = tf.matmul(do2, fc3_W) + fc3_b
    fc3 = tf.nn.relu(fc3) 
    
#     fc4_W = tf.Variable(tf.truncated_normal(shape=(256,128), mean=mu, stddev=sigma))
#     fc4_b = tf.Variable(tf.zeros(128))
#     fc4 = tf.matmul(fc3, fc4_W) + fc4_b
#     fc4 = tf.nn.relu(fc4) 
    do3 = tf.nn.dropout(fc3, 0.5)
    
    fc5_W = tf.Variable(tf.truncated_normal(shape=(128,n_classes), mean=mu, stddev=sigma))
    fc5_b = tf.Variable(tf.zeros(n_classes))
    logits = tf.matmul(do3, fc5_W) + fc5_b
    
    return logits
```

```python
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)
```

### Train, Validate and Test the Model

I used **cross entropy** as my model's loss function and initial the **learning rate** to **5e-4** (normal setting for `AdamOptimizer`). 

And `BATCH_SIZE` was 64, which was also a common choice. I haved tried to set BATCH_SIZE bigger to make the training faster but the learning rate was hard to configure.

```python
lr = 5e-4
BATCH_SIZE = 64

logits = LeNet(x)
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=one_hot_y))

optimizer = tf.train.AdamOptimizer(learning_rate=lr)
training_op = optimizer.minimize(loss)
```

```python
correct_prediction = tf.equal(tf.argmax(logits,1), tf.argmax(one_hot_y, 1))
accuracy_op = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_acc = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        acc = sess.run(accuracy_op, feed_dict={x:batch_x, y:batch_y})
        total_acc += (acc * len(batch_x))
        
    return total_acc / num_examples
```

```python
EPOCHS = 0

with tf.Session() as sess:
    # Train from zero
    sess.run(tf.global_variables_initializer())
    
    # Train from saved models
#     saver = tf.train.Saver()
#     saver.restore(sess, './saved_models/Official_Model_4_0.939683')

    num_examples = len(X_train_normed)
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train_normed, y_train = shuffle(X_train_normed, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train_normed[offset:end], y_train[offset:end]
            sess.run(training_op, feed_dict={x:batch_x, y:batch_y})
            
        validation_acc = evaluate(X_valid_normed, y_valid)
        print("EPOCH {} ...".format(i))
        print("Validation Accuracy = {:.5f}".format(validation_acc))
        print()

        if validation_acc >= 0.93:
            try:
                saver
            except NameError:
                saver = tf.train.Saver()
            saver.save(sess, "./saved_models/Official_Model_{}_{:.6f}".format(i, validation_acc))
            print("Model {} saved".format(i))
    print("Congratulations! Training was finished!")
```

### Solution Approach
#### Correct pre-processing was really important
I applied the 0-mean and 1-std to each pixels and it made the data untrainable for some of the images are too dark and some are too bright. CLAHE was good for this task.

#### Learning rate was essential  
With learning rate of 5e-4, it took over 330 epochs to get to the best. Obviously, with a lower learning rate, it might be much longer than that. But in contrast, if the learning rate was too large, then the training would be unstable and sometimes likely to step back to the initial point.

#### Baseline gave me confidence  
With using a good pre-processing and a good learning rate, I got a 82% accuracy by the `LeNet` (about 200 epochs). And I was clear that I still had some useful algorithms to enhance that. First I tried to built a deeper neural network, but training was slow, so I applied another useful algorithms named `dropout` to speed it up.

#### Decide the final structure  
I was not training hundreds of epochs after each modification applied to the network. I watched the accuracy of the head 50 epoch, if it slow down at about 70%, it might probably not be a nice shot. For the first time I noticed the accuracy rised slow down at 83%, I thought the framework of the network structure was correct, so I tried a few different learning rate. Finally, I was managed to get 93% accuracy at validation set.

```python
with tf.Session() as sess:
    saver = tf.train.Saver()
    saver.restore(sess, './saved_models/Official_Model_4_0.939683')
    eval_train = evaluate(X_train_normed, y_train)
    print("Train Set Accuracy is {}".format(eval_train))
    eval_valid = evaluate(X_valid_normed, y_valid)
    print("Validation Set Accuracy is {}".format(eval_valid))
    eval_test = evaluate(X_test_normed, y_test)
    print("Test Set Accuracy is {}".format(eval_test))
```

```python
import pandas as pd
import matplotlib.image as mpimg
import glob

names = pd.read_csv('./signnames.csv')
names = np.array(names.iloc[:, 1])

test_files = glob.glob('./test_images/*')

f, (ax) = plt.subplots(2, 3, figsize=(24, 12))

for ind, file in enumerate(test_files):
    img = mpimg.imread(file)
    ax[np.int(ind // 3)][np.int(ind % 3)].imshow(img)
```

### Predict the Sign Type for Each Image

```python
test_five = []

for file in test_files:
    img = mpimg.imread(file)
    test_five.append(img)
    
test_five = np.array(test_five)
test_five_normed = test_five.reshape(-1, 32*32*3)
test_five_normed = np.apply_along_axis(CLAHE, 1, test_five_normed, (3,3)).astype(np.float32).reshape(-1, 32, 32, 1)

pred_classes = None
pred_logits = None

with tf.Session() as sess:
    saver = tf.train.Saver()
    saver.restore(sess, './saved_models/Official_Model_4_0.939683')
    
    pred_logits = sess.run(logits, feed_dict={x:test_five_normed})
    pred_classes = np.argmax(pred_logits, axis=1)
    f, (ax) = plt.subplots(2, 3, figsize=(24, 12))

    for ind, file in enumerate(test_five):
        ax_tmp = ax[np.int(ind // 3)][np.int(ind % 3)]
        ax_tmp.set_title(names[pred_classes[ind]], size=25)
        ax_tmp.imshow(test_five[ind])
```
